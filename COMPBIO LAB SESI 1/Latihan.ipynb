```python
import Bio
from Bio.Seq import Seq
from Bio.SeqUtils import nt_search
```

```python
seqA = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
seqB = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

```python
print(len(seqA))
print(len(seqB))
```

```python
# nt_search(str(seqA), 'CGC')
# nt_search(str(seqB), 'CGC')

countA = seqA.count('GC')
countB = seqB.count('GC')
print(countA)
print(countB)
```

```python
idxA = seqA.find('CAGTC')
idxB = seqB.find('CAGTC')

print(idxA)
print(idxB)
```
