```python
# Pip install biopython - Collab
import Bio
from Bio.Seq import Seq
from Bio.SeqUtils import nt_search
```

```python
# Create Sequence, Length, Slicing, Reverse, Concatenating, Find, Index, Find Substring
seq1 = Seq('ATGCGCTAGT')
seq1
```

```python
# Length
len(seq1)        
```

```python
# Slicing
# start:stop:step     
# ATGCGCTAGT
print(seq1[0:10:2])
print(seq1[::-1])
seq1[::-1]
```

```python
# Concatenating
seq2 = Seq('ATTGCT')
seq3 = seq1 + seq2
seq3
```

```python
# Find
# ATGCGCTAGT
seq1.find('AG')
```

```python
# Index
# ATGCGCTAGT
seq1.index('GC')
```

```python
# Find Substring
# ATGCGCTAGT
nt_search(str(seq1), 'T')

# return subs, index ia ditemukan
# dir(Seq)
```

```python
# Plot Nucleotide Frequency in DNA Sequence
# Counter
from collections import Counter

counter_seq = Counter(seq3)
counter_seq
```

```python
from matplotlib import pyplot as plt
# Counter({'T': 6, 'G': 4, 'A': 3, 'C': 3})
plt.bar(counter_seq.keys(), counter_seq.values())
```
