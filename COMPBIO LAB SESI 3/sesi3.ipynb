```python
from Bio.Seq import Seq
```

```python
# DNA -> coding strand (sense strand)
dna = Seq('ATGATCTCGTAA')
dna
```

```python
# cDNA -> template strand (antisense strand)
cdna = dna.reverse_complement()
cdna
```

```python
# Transcribe : DNA -> mRNA
mrna = dna.transcribe()
mrna
```

```python
# Translate -> mRNA -> polypeptide (amino acids)
polypeptide = mrna.translate()
polypeptide

# = Seq('MIS*') * -> stop codon
```

```python
from Bio.SeqUtils import seq1, seq3
```

```python
polypeptide = seq3(polypeptide)
polypeptide
# = MetIleSerTer
```

```python
seq1(polypeptide)
```
