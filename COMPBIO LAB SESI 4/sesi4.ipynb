```python
from Bio.Seq import Seq
from Bio import pairwise2
from Bio.pairwise2 import format_alignment as fa
from Levenshtein import distance
```

```python
seq1 = Seq("CGATGACAAGGCATGACGATACGATAATAGCATGCA")
seq2 = Seq("ATGACGAAATGACTAAGCAAAGCAATGCA")
```

```python
global_alignment = pairwise2.align.globalxx(seq1, seq2)
for a in global_alignment:
    print(fa(*a))
```

```python
local_alignment= pairwise2.align.localxx(seq1, seq2)
for a in local_alignment:
    print(fa(*a))
```

```python
match_score = +2
mismatch_score = -1
gap_open = -0.5
gap_extend = -0.1

gA_match = pairwise2.align.globalms(seq1, seq2,
                                    match_score,
                                    mismatch_score,
                                    gap_open, gap_extend)

for a in gA_match:
    print(fa(*a))
```

```python
def hamming_distance(s1, s2):
    differences = []
    for x, y in zip(s1, s2):
        if x != y:
            differences.append((x, y))
    return len(differences)

print(f"Hamming: {hamming_distance(seq1, seq2)}")
```

```python
epstein_dist = distance(seq1, seq2)
print(f"Minimum operation: {epstein_dist}")
```

```python
def dotplot(s1, s2, nonblank=chr(0x25A0)):
    print(" | " + s2)
    print("-" * (len(s2) + 2))
    
    for x in s1:
        line = ''
        for y in s2:
            if x == y:
                line += nonblank
            else:
                line += ' '     
        print(x + "|" + line)

seq1 = "CGATGACAAGGCATGACGATACGATAATAGCATGCA"
seq2 = "ATGACGAAATGACTAAGCAAAGCAATGCA"
dotplot(seq1, seq2)
```
