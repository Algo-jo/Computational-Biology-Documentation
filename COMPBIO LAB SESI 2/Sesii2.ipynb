```python
from Bio.Seq import Seq
```

```python
seq = Seq('GATATCGTAATCGGCATGCTA')
print(seq)
```

```python
from Bio.SeqUtils import gc_fraction
```

```python
gc = gc_fraction(seq)
print(gc)
# (count g + count c) / len

print(f"gc content sq: {gc*100}%\n")

at = 1 - gc
print(at)
print(f"at content: {at*100}%\n")
```

```python
# Hitung Manual
def hitung_gc(seq):
    count_G = seq.count('G')
    count_C = seq.count('C')
    len_seq = len(seq)
    return (count_G + count_C) / len_seq

print(hitung_gc(seq))
```

```python
from Bio.SeqUtils import MeltingTemp as mt
```

```python
# TM -> suhu saat 50% DNA double-stranded berpisah menjadi single-stranded
# Metode Wallace -> filtering awal, cepat & kasar
mt_wallace = mt.Tm_Wallace(seq)
print(f"mt wallace: {mt_wallace}")

# Metode GC -> Sekuens sedang, estimasi umum
mt_GC = mt.Tm_GC(seq)
print(f"mt GC: {mt_GC:.2f}")

# Metode Nearest Neighbour (NN) -> sekuens panjang, paling akurat
mt_NN = mt.Tm_NN(seq)
print(f"mt NN: {mt_NN:.2f}")
```
