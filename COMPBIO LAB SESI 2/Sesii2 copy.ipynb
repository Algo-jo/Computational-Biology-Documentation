```python
from Bio.Seq import Seq
```

```python
seqA = Seq('GCGCGCGCGCGC')
seqB = Seq('ATGATCGTAATCGGCATGCTA')
seqC = Seq('TTAATTAATTAA')
```

```python
from Bio.SeqUtils import gc_fraction
```

```python
def gc_fractionX(seq, label):
    gc = gc_fraction(seq)
    print(f"gc content sq: {gc*100:.2f}%")
    at = 1 - gc
    print(f"at content {label}: {at*100:.2f}%")

gc_fractionX(seqA, "A")
gc_fractionX(seqB, "B")
gc_fractionX(seqC, "C")
```

```python
from Bio.SeqUtils import MeltingTemp as mt
```

```python
# FRAGMENT A
print("=== Fragment A ===")
mt_wallaceA = mt.Tm_Wallace(seqA)
mt_GCA = mt.Tm_GC(seqA)
mt_NNA = mt.Tm_NN(seqA)
print(f"mt wallace A: {mt_wallaceA:.2f}")
print(f"mt GC A: {mt_GCA:.2f}")
print(f"mt NN A: {mt_NNA:.2f}")

# FRAGMENT B
print("=== Fragment B ===")
mt_wallaceB = mt.Tm_Wallace(seqB)
mt_GCB = mt.Tm_GC(seqB)
mt_NNB = mt.Tm_NN(seqB)
print(f"mt wallace B: {mt_wallaceB:.2f}")
print(f"mt GC B: {mt_GCB:.2f}")
print(f"mt NN B: {mt_NNB:.2f}")

# FRAGMENT C
print("=== Fragment C ===")
mt_wallaceC = mt.Tm_Wallace(seqC)
mt_GCC = mt.Tm_GC(seqC)
mt_NNC = mt.Tm_NN(seqC)
print(f"mt wallace C: {mt_wallaceC:.2f}")
print(f"mt GC C: {mt_GCC:.2f}")
print(f"mt NN C: {mt_NNC:.2f}")
```

```python
from Bio.SeqUtils import molecular_weight as mw
```

```python
# Hitung DNA
mw_seqA_DNA = mw(seqA)
mw_seqB_DNA = mw(seqB)
mw_seqC_DNA = mw(seqC)
print(f"Molecular weight (DNA): {mw_seqA_DNA:2f}")
print(f"Molecular weight (DNA): {mw_seqB_DNA:.2f}")
print(f"Molecular weight (DNA): {mw_seqC_DNA:.2f}")
```

```python
# Task 4 — Fungsi Reusable
# Bungkus semua analisis di atas ke dalam satu fungsi bernama analyze_dna(seq, label nama) yang menerima objek Seq dan nama label-nya, lalu mencetak semua hasil (GC%, AT%, melting temperature, dan Molecular Weight) dalam satu blok output yang rapi. Panggil fungsi tersebut untuk salah satu fragmen.
```

```python
def analyze_dna(seq, label):
    gc = gc_fraction(seq)
    at = 1 - gc
    mt_wallace = mt.Tm_Wallace(seq)
    mt_GC = mt.Tm_GC(seq)
    mt_NN = mt.Tm_NN(seq)
    mw_seq_DNA = mw(seq)
    
    print(f"==={label}===")
    print(f"gc content sq: {gc*100:.2f}%")
    print(f"at content: {at*100:.2f}%")
    print(f"mt wallace: {mt_wallace:.2f}")
    print(f"mt GC: {mt_GC:.2f}")
    print(f"mt NN: {mt_NN:.2f}")
    print(f"Molecular weight (DNA): {mw_seq_DNA:2f}")

analyze_dna(seqA, "Fragment A")
```
